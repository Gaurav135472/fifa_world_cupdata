```python
def normalize_team_name(team):
    return team.lower().strip()

# Define the function to determine the stage of the match
def assign_tournament_stage(row):
    match_year = str(row['Date'].year)  # Extract the year from the date
    
    # Check if the match year exists in the dictionary
    if match_year in match_stages:
        match_date = row['Date'].strftime('%Y-%m-%d')
        
        # Check if the match date exists for that year
        if match_date in match_stages[match_year]:
            # Normalize the teams for matching
            home_team = normalize_team_name(row['Home_Team'])
            away_team = normalize_team_name(row['Away_Team'])
            
            # Iterate over all matches on that date to find the matching teams
            for teams in match_stages[match_year][match_date]['teams']:
                # Normalize the teams from the dictionary
                team1 = normalize_team_name(teams[0])
                team2 = normalize_team_name(teams[1])
                
                # Match the teams regardless of the order
                if {team1, team2} == {home_team, away_team}:
                    return match_stages[match_year][match_date]['stage']
    
    # Default to 'Group Stage' if no match is found
    return 'Group Stage'

# Convert 'Date' column to datetime if it's not already
df['Date'] = pd.to_datetime(df['Date'])

# Apply the function to create a new column for the tournament stage
df['Tournament_Stage'] = df.apply(assign_tournament_stage, axis=1)

# Verify the result
print(df[['Date', 'Home_Team', 'Away_Team', 'Tournament_Stage']].head())
```
